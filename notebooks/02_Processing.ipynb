# **3.1.	Kiểm chứng nghịch lý trong ngành CS**

```python
import pandas as pd
from typing import Tuple
from pathlib import Path
import sys
```

```python
project_root = Path.cwd().parent
sys.path.insert(0, str(project_root))

from src.load_data import load_data, save_data
from src.processing import clean_data, filter_cs_occupations
```

```python
# 1. Load dữ liệu gốc
df_desires = load_data("domain_worker_desires.csv")
df_metadata = load_data("domain_worker_metadata.csv")
df_task = load_data("task_statement_with_metadata.csv")
df_expert = load_data("expert_rated_technological_capability.csv")
```

```python
# 2. Xử lý làm sạch
df_desires_clean, df_metadata_clean = clean_data(df_desires, df_metadata)
```

```python
# 3. Lưu file đã làm sạch
path_desires = save_data(df_desires_clean, "df_desires_clean.csv")
path_metadata = save_data(df_metadata_clean, "df_metadata_clean.csv")
```

```python
# 4. Dùng file mới để lọc (Đọc lại từ file vừa lưu)
df_desires_new = load_data("df_desires_clean.csv", file_path=path_desires)
df_metadata_new = load_data("df_metadata_clean.csv", file_path=path_metadata)
```

```python
# 5. Lọc ngành CS
cs_desires, cs_metadata, cs_expert, cs_occupations_df, cs_task = filter_cs_occupations(
    df_task, df_desires_new, df_metadata_new, df_expert
)
```

```python
# 6. Lưu kết quả cuối cùng
save_data(cs_desires, "cs_desires.csv")
save_data(cs_metadata, "cs_metadata.csv")
save_data(cs_expert, "cs_expert.csv")
save_data(cs_task, "cs_task.csv")
save_data(cs_occupations_df, "cs_occupations.csv")

print("Hoàn tất quy trình: Làm sạch -> Lưu -> Đọc lại -> Lọc!")
```
